# FLEKS Python Visualization Toolkit

flekspy is a Python package for processing FLEKS data.

## FLEKS data format

* Field: *.out format or AMREX built-in format, whose directory name is assumed to end with "_amrex"
* PIC particle: AMREX built-in format 
* Test particle: binary data format

## Visualizing fields

We provide convenient `pp` and `pcolormesh` method for IDL 1D and 2D outputs:

```python
file = "data/1d__raw_2_t25.60000_n00000258.out"
ds = flekspy.load(file)
ds.plot("p", "Bx")
```

```python
file = "data/z=0_fluid_region0_0_t00001640_n00010142.out"
ds = flekspy.load(file)
ds.pcolormesh("x", "y", "Bx", "By", "Bz", scale=False)
```

Native AMReX format can be read using YT:

```python
ds = flekspy.load("data/3d*amrex")
```

We can take a slice at a given location, which works both for IDL and AMReX data:

```python
dc = ds.get_slice("z", 0.5)
dc
```

and plot the selected variables on the slice with colored contours, streamlines, and contour lines:

```python
f, axes = dc.plot("By", pcolor=True)
dc.add_stream(axes[0], "Bx", "By", color="w")
dc.add_contour(axes[0], "Bx")
```

Users can also obtain the data of a 2D plane, and visualize it with matplotlib. A complete example is given below:

```python
import matplotlib.pyplot as plt
import numpy as np

ds = flekspy.load("data/3*amrex")
dc = ds.get_slice("z", 0.5)

f, axes = plt.subplots(
    1, 2, figsize=(12, 4), constrained_layout=True, sharex=True, sharey=True
)

fields = ["By", "Bz"]
for ivar in range(2):
    v = dc.evaluate_expression(fields[ivar])
    vmin = v.min().v
    vmax = v.max().v
    ax = axes[ivar]
    ax.set_title(fields[ivar], fontsize=16)
    ax.set_ylabel("Y", fontsize=16)
    ax.set_xlabel("X", fontsize=16)
    c = ax.pcolormesh(dc.x.value, dc.y.value, np.array(v.T), cmap="turbo")
    cb = f.colorbar(c, ax=ax, pad=0.01)

    ax.set_xlim(np.min(dc.x.value), np.max(dc.x.value))
    ax.set_xlim(np.min(dc.y.value), np.max(dc.y.value))
    dc.add_stream(ax, "Bx", "By", density=0.5, color="w")

plt.show()
```

## Visualizing phase space distributions

```python
from flekspy.util import unit_one

data_file = "data/3d_*amrex"
ds = flekspy.load(data_file)

# Add a user defined field. See yt document for more information about derived field.
ds.add_field(
    ds.pvar("unit_one"),
    function=unit_one,
    sampling_type="particle",
    units="dimensionless",
)

x_field = "p_uy"
y_field = "p_uz"
# z_field = "unit_one"
z_field = "p_w"
xleft = [-0.016, -0.01, 0.0]
xright = [0.016, 0.01, 1.0]

zmin, zmax = 1e-5, 2e-3

## Select and plot the particles inside a box defined by xleft and xright
region = ds.box(xleft, xright)
pp = ds.plot_phase(
    x_field,
    y_field,
    z_field,
    region=region,
    unit_type="si",
    x_bins=100,
    y_bins=32,
    domain_size=(xleft[0], xright[0], xleft[1], xright[1]),
)

pp.set_cmap(pp.fields[0], "turbo")
pp.set_font(
    {
        "size": 34,
        "family": "DejaVu Sans",
    }
)

pp.set_zlim(pp.fields[0], zmin=zmin, zmax=zmax)

pp.set_xlabel(r"$V_y$")
pp.set_ylabel(r"$V_z$")
pp.set_colorbar_label(pp.fields[0], "weight")
str_title = (
    rf"x = [{xleft[0]:.1e}, {xright[0]:.1e}], y = [{xleft[1]:.1e}, {xright[1]:.1e}]"
)
pp.set_title(pp.fields[0], str_title)
pp.set_log(pp.fields[0], True)
pp.show()
# pp.save("test")
# pp.plots[("particles", z_field)].axes.xaxis.set_major_locator(
# ticker.MaxNLocator(4))
# pp.plots[("particles", z_field)].axes.yaxis.set_major_locator(
# ticker.MaxNLocator(4))
```

```python
import numpy as np

def calc_phase_space_dis(region, var_x, var_y):
    """
    Calculate the phase space distribution of particles in a given region.
    
    Parameters:
    - region: The region of interest.
    - var_x: Tuple specifying the x variable, e.g. ('particles', 'particle_position_x')
    - var_y: Tuple specifying the y variable, e.g. ('particles', 'particle_velocity_x')
    
    Returns:
    - A 2D histogram tuple (hist, xedges, yedges)
    """
    #data_x = region[var_x] unitless
    #data_y = region[var_y] unitless
    data_x = region[var_x].to_value() # unitless
    data_y = region[var_y].to_value() # unitless

    hist2d, xedges, yedges = np.histogram2d(data_x, data_y, bins=100, range=[[-1, 1], [-1, 1]])
    
    return hist2d, xedges, yedges
```

```python
import matplotlib.pyplot as plt

def plot_plot_phase(hist2d_tuple):
    """
    Plot a 2D histogram (phase-space distribution).
    
    Parameters:
    - hist2d_tuple: A tuple of (hist, xedges, yedges) from np.histogram2d
    """
    hist, xedges, yedges = hist2d_tuple
    
    plt.figure(figsize=(6, 5))
    plt.imshow(hist.T, origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
               aspect='auto', cmap='viridis')
    plt.colorbar(label='Counts')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Phase Space Distribution')
    plt.tight_layout()
    plt.show()
```

```python
var_x = ('particles', 'particle_velocity_x')
var_y = ('particles', 'particle_velocity_y')
```

```python
hist2d = calc_phase_space_dis(region, var_x, var_y)
plot_plot_phase(hist2d)
```

```python
def maxwellian_2d(vx, vy, A, vx0, vy0, sigma_x, sigma_y):
    return A * np.exp(
        -((vx - vx0)**2 / (2 * sigma_x**2) + (vy - vy0)**2 / (2 * sigma_y**2))
    )
```

```python
hist, xedges, yedges = calc_phase_space_dis(region, 
        ('particles', 'particle_velocity_x'),
        ('particles', 'particle_velocity_y'))

# Bin centers
x_centers = 0.5 * (xedges[:-1] + xedges[1:])
y_centers = 0.5 * (yedges[:-1] + yedges[1:])
X, Y = np.meshgrid(x_centers, y_centers, indexing='ij')

# Example: manually choose parameters
A = np.max(hist)
vx0 = 0.0
vy0 = 0.0
sigma_x = 0.3
sigma_y = 0.3

M = maxwellian_2d(X, Y, A, vx0, vy0, sigma_x, sigma_y)

import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot histogram
axes[0].imshow(hist.T, origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], aspect='auto')
axes[0].set_title("Particle Phase Space Histogram")
axes[0].set_xlabel("vx")
axes[0].set_ylabel("vy")

# Plot Maxwellian
axes[1].imshow(M.T, origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], aspect='auto')
axes[1].set_title("Analytical Maxwellian")
axes[1].set_xlabel("vx")
axes[1].set_ylabel("vy")

plt.tight_layout()
plt.show()

```

### Selecting particles starting in a region

```python
def f_select(tp, pid):
    pData = tp.read_initial_location(pid)
    inRegion = pData[FLEKSTP.ix_] > 0 and pData[FLEKSTP.iy_] > 0
    return inRegion


pSelected = tp.select_particles(f_select)
```
